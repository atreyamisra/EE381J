```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def multiplicationOfUniforms(n):
    U=np.random.uniform(0,2,10000)
    for i in xrange(n):
        Un=np.random.uniform(0,2,10000)
        U=U*Un
    return U
Xn=multiplicationOfUniforms(1)
plt.hist(Xn, bins=100)
plt.show()
plt.close()
Xn=multiplicationOfUniforms(4)
plt.hist(Xn, bins=100)
plt.show()
plt.close()
Xn=multiplicationOfUniforms(10)
plt.hist(Xn, bins=100)
plt.show()
plt.close()
Xn=multiplicationOfUniforms(100)
plt.hist(Xn, bins=100)
plt.show()
plt.close()
```

```python
def sinXn(n):
    w=np.random.uniform(0,1,10000)
    Xn=w*np.sin(2*np.pi*n*w)
    return Xn
X=sinXn(1)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=sinXn(10)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=sinXn(100)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=sinXn(100000)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=sinXn(1000000000000000000000000)
plt.hist(X, bins=100)
plt.show()
plt.close()
```

```python
def productExponent(n):
    w=np.random.uniform(0,1,10000)
    return n*(w)**n
X=productExponent(1)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=productExponent(10)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=productExponent(1000)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=productExponent(10000000000)
plt.hist(X, bins=100)
plt.show()
plt.close()
```

```python
def subtractQuantizer(n):
    w=np.random.uniform(0,1,10000)
    return n*w-np.floor(n*w)
X=subtractQuantizer(1)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=subtractQuantizer(100)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=subtractQuantizer(1000)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=subtractQuantizer(10000000)
plt.hist(X, bins=100)
plt.show()
plt.close()
```

```python
def nSquaredOneOverN(n):
    w=np.random.uniform(0,1/n, 10000)
    Xn=(n**2)*w
    return Xn
X=nSquaredOneOverN(1)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=nSquaredOneOverN(10)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=nSquaredOneOverN(100)
plt.hist(X, bins=100)
plt.show()
plt.close()
X=nSquaredOneOverN(10000)
plt.hist(X, bins=100)
plt.show()
plt.close()
```
